## cleaning data NA

```python
df = df.dropna()
```

## mencari data yg duplicat

```python
df.duplicated()
```

## mensama ratakan semua str yg besar dan tidak rapih menjadi sama semua menjadi lower

```python
df['Artist'] = df['Artist'].str.lower()
df['Track'] = df['Track'].str.lower()
df['Album'] = df['Album'].str.lower()
df['Title'] = df['Title'].str.lower()
df['Channel'] = df['Channel'].str.lower()
```

```python
df
```

## Views Like Stream Paling banyak

```python
most_views_Likes_stream = df.sort_values(by=['Views','Likes','Stream'], ascending=False)
```
